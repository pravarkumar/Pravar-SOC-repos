```python
class PatchEmbedding(nn.Module):
    def __init__(self, img_size=32, patch_size=4, in_channels=3, embed_dim=128):
        super().__init__()#take 4X4 and make it to 128
        self.num_patches = (img_size // patch_size) ** 2  # = 64 patches
        # this one layer does: cut into patches + flatten + linear projection
        self.proj = nn.Conv2d(in_channels, embed_dim, kernel_size=patch_size, stride=patch_size)
#Take a 4X4 path look at its 48 numbers and convert thme into 128 numbers .
    #each path is (3,4,4) so 48 numbers 
    def forward(self, x):
        x = self.proj(x)       # shape: (batch, embed_dim, 8, 8)
        x = x.flatten(2)       # shape: (batch, embed_dim, 64)
        x = x.transpose(1, 2)  # shape: (batch, 64, embed_dim)
        return x

# quick test
pe = PatchEmbedding()
dummy = torch.randn(2, 3, 32, 32)  # 2 images, 3 channels, 32x32
out = pe(dummy)# same as out = pe.forward(dummy)
print(out.shape)  # should print: torch.Size([2, 64, 128]) 

#out on going to forward becomes (2,128,8,8) on flattening:(2,123,64) on transposing (2,64,128)
#gives us(64,128) a vector for a set of 64 patches we input in it.
```

```python
class ViT(nn.Module):
    def __init__(self, img_size=32, patch_size=4, in_channels=3,
                 embed_dim=128, depth=4, num_heads=4, mlp_dim=256, num_classes=10):
        super().__init__()

        self.patch_embed = PatchEmbedding(img_size, patch_size, in_channels, embed_dim)
        num_patches = (img_size // patch_size) ** 2  # 64

        # CLS token: a learnable vector prepended to the sequence
        # think of it as "summary slot" — after attention, it summarizes the whole image
        self.cls_token = nn.Parameter(torch.randn(1, 1, embed_dim))#(1,1,128)#we are making token using random numberslearn this during training

        # positional embedding: tells the model where each patch came from
        # +1 because of the CLS token
        #the first step or any model is to make it random then teach it foundation is made from random only
        self.pos_embed = nn.Parameter(torch.randn(1, num_patches + 1, embed_dim))#(1,65,128) one space reserve for the type of imageit is 

        # the transformer encoder: this is where self-attention(tokens talkign to each other) happens
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=embed_dim, nhead=num_heads,
            dim_feedforward=mlp_dim, batch_first=True
        )
        self.transformer = nn.TransformerEncoder(encoder_layer, num_layers=depth)
#It allows every patch token and the CLS token
#to exchange information using self-attention.
#Multiple transformer layers repeatedly refine
#these representations until the CLS token
#contains information about the whole image.
        self.norm = nn.LayerNorm(embed_dim)

        # final classifier: takes CLS token → 10 class scores
        self.head = nn.Linear(embed_dim, num_classes)

    def forward(self, x):
        B = x.shape[0]  # batch size

        # step 1: image → 64 patch vectors
        x = self.patch_embed(x)

        # step 2: prepend CLS token to every image in the batch
        cls = self.cls_token.expand(B, -1, -1)  # (B, 1, 128)
        x = torch.cat([cls, x], dim=1)          # (B, 65, 128)

        # step 3: add position info
        x = x + self.pos_embed

        # step 4: run through transformer (self-attention happens here)
        x = self.transformer(x)
        x = self.norm(x)#we sacle all the numbers to be comparable

        # step 5: take only the CLS token output → classify
        cls_out = x[:, 0]           # (B, 128)
        return self.head(cls_out)   # (B, 10)

# quick test
model = ViT()
dummy = torch.randn(2, 3, 32, 32)
out = model(dummy)
print(out.shape)  # should be torch.Size([2, 10])
print(f"Total parameters: {sum(p.numel() for p in model.parameters()):,}")
```

```python
device = torch.device("mps" if torch.backends.mps.is_available() else "cpu")
print(f"Using: {device}")

model = ViT().to(device)
optimizer = torch.optim.AdamW(model.parameters(), lr=3e-4)
criterion = nn.CrossEntropyLoss()

def train_one_epoch(epoch):
    model.train()
    total_loss, correct, total = 0, 0, 0

    for i, (images, labels) in enumerate(trainloader):
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()        # clear old gradients
        outputs = model(images)      # forward pass
        loss = crit erion(outputs, labels)  # compare to correct labels
        loss.backward()              # compute gradients
        optimizer.step()             # update weights

        total_loss += loss.item()
        predicted = outputs.argmax(dim=1)  # highest score = model's guess
        correct += (predicted == labels).sum().item()
        total += labels.size(0)

        if i % 100 == 0:
            print(f"Epoch {epoch} | Step {i}/{len(trainloader)} | Loss: {loss.item():.3f}")

    acc = 100 * correct / total
    print(f"Epoch {epoch} done | Avg Loss: {total_loss/len(trainloader):.3f} | Train Acc: {acc:.1f}%")
    return acc

def evaluate():
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():  # no gradients needed for evaluation
        for images, labels in testloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            predicted = outputs.argmax(dim=1)
            correct += (predicted == labels).sum().item()
            total += labels.size(0)
    acc = 100 * correct / total
    print(f"Test Accuracy: {acc:.1f}%")
    return acc

# train for 10 epochs (enough to see it learning, not too slow)
for epoch in range(1, 11):
    train_one_epoch(epoch)
    evaluate()
```

```python
class VICReg(nn.Module):
    def __init__(self, embed_dim=128, proj_dim=64):
        super().__init__()
        
        # encoder: our ViT but WITHOUT the classifier head
        self.encoder = ViT(num_classes=embed_dim)  # outputs 128-dim vector instead of 10 classes
        
        # projector: small 2-layer MLP on top of encoder
        # this is where VICReg loss is applied
        self.projector = nn.Sequential(
            nn.Linear(embed_dim, embed_dim),
            nn.ReLU(),
            nn.Linear(embed_dim, proj_dim)  # final 64-dim projection
        )
    
    def forward(self, x):
        return self.projector(self.encoder(x))

# quick test
vicreg_model = VICReg().to(device)
dummy = torch.randn(2, 3, 32, 32).to(device)
out = vicreg_model(dummy)
print(f"Output shape: {out.shape}")  # should be (2, 64)
print(f"Parameters: {sum(p.numel() for p in vicreg_model.parameters()):,}")
```

```python
def vicreg_loss(z1, z2, lam=25, mu=25, nu=1):
    N, D = z1.shape  # N = batch size, D = projection dim (64)
    
    # I — Invariance: two views of same image should have similar embeddings
    invariance = nn.functional.mse_loss(z1, z2)
    
    # V — Variance: each dimension should have std >= 1 across the batch
    # hinge loss: only penalizes if std drops below 1
    std_z1 = torch.sqrt(z1.var(dim=0) + 1e-4)
    std_z2 = torch.sqrt(z2.var(dim=0) + 1e-4)
    variance = torch.mean(nn.functional.relu(1 - std_z1)) + \
               torch.mean(nn.functional.relu(1 - std_z2))
    
    # C — Covariance: off-diagonal terms of covariance matrix should be 0
    # stops dimensions encoding the same thing
    z1 = z1 - z1.mean(dim=0)  # center the embeddings
    z2 = z2 - z2.mean(dim=0)
    cov_z1 = (z1.T @ z1) / (N - 1)  # DxD covariance matrix
    cov_z2 = (z2.T @ z2) / (N - 1)
    # penalize off-diagonal terms only
    off_diag_z1 = cov_z1 - torch.diag(torch.diag(cov_z1))
    off_diag_z2 = cov_z2 - torch.diag(torch.diag(cov_z2))
    covariance = off_diag_z1.pow(2).sum() / D + \
                 off_diag_z2.pow(2).sum() / D
    
    # total loss
    loss = lam * invariance + mu * variance + nu * covariance
    return loss, invariance.item(), variance.item(), covariance.item()

# quick test
z1 = torch.randn(64, 64)
z2 = torch.randn(64, 64)
loss, i, v, c = vicreg_loss(z1, z2)
print(f"Total loss: {loss:.3f} | I: {i:.3f} | V: {v:.3f} | C: {c:.3f}")
```

```python
vicreg_model = VICReg().to(device)
optimizer = torch.optim.AdamW(vicreg_model.parameters(), lr=3e-4)

def train_vicreg(epochs=10):
    for epoch in range(1, epochs + 1):
        vicreg_model.train()
        total_loss = 0
        
        for i, (view1, view2, _) in enumerate(trainloader_2view):
            # notice: we ignore labels completely (_)
            view1, view2 = view1.to(device), view2.to(device)
            
            z1 = vicreg_model(view1)  # embed view 1
            z2 = vicreg_model(view2)  # embed view 2
            
            loss, inv, var, cov = vicreg_loss(z1, z2)
            
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            
            total_loss += loss.item()
            
            if i % 100 == 0:
                print(f"Epoch {epoch} | Step {i}/{len(trainloader_2view)} | Loss: {loss.item():.3f} | I: {inv:.3f} | V: {var:.3f} | C: {cov:.3f}")
        
        print(f"Epoch {epoch} done | Avg Loss: {total_loss/len(trainloader_2view):.3f}\n")

train_vicreg(epochs=10)
```

```python
class SmallProjectorVICReg(nn.Module):
    def __init__(self, embed_dim=128, proj_dim=64):
        super().__init__()

        self.encoder = ViT(num_classes=embed_dim)

        self.projector = nn.Linear(
            embed_dim,
            proj_dim
        )

    def forward(self, x):
        return self.projector(
            self.encoder(x)
        )
```

```python
small_model = SmallProjectorVICReg().to(device)

optimizer = torch.optim.AdamW(
    small_model.parameters(),
    lr=3e-4
)

for epoch in range(3):

    small_model.train()

    for view1, view2, _ in trainloader_2view:

        view1 = view1.to(device)
        view2 = view2.to(device)

        z1 = small_model(view1)
        z2 = small_model(view2)

        loss, _, _, _ = vicreg_loss(z1, z2)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    print("Epoch", epoch + 1, "done")
```
